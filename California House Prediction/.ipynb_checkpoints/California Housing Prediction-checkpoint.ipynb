## Scaling Numeric Features

Another good practice is to scale numeric features to a small range of values e.g. $(0,1)$ or $(-1,1)$. Scaling numeric features ensures that no particular feature has a disproportionate impact on the model's loss. Optimization algorithms also work better in practice with smaller numbers.

The numeric columns in our dataset have varying ranges.

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
scalar.fit(house_df[numeric_cols])
```

```python
scalar.data_max_
```

```python
scalar.data_min_
```

```python
house_df[numeric_cols]=scalar.transform(house_df[numeric_cols])
```

```python
house_df.describe()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse=False,handle_unknown='ignore')
```

```python
encoder.fit(house_df[categoraial_cols])
```

```python
encoder.categories_
```

```python
encoded_cols=list(encoder.get_feature_names_out(categoraial_cols))
print(encoded_cols)
```

```python
house_df[encoded_cols]=encoder.transform(house_df[categoraial_cols])
```
